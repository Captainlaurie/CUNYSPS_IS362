```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

#import data into pandas dataframe and specify column names
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/auto-mpg/auto-mpg.data"
column_names = ['MPG', 'Cylinders', 'Displacement', 'Horsepower', 'Weight', 'Acceleration', 'Model Year', 'Origin', 'Car Name']

auto_data_df = pd.read_csv(url, delim_whitespace=True, names=column_names)

#Check the data
auto_data_df.head()

```

```python
#Change missing Horsepower data to NaN, and change the Horsepower column type to float
auto_data_df['Horsepower'] = auto_data_df['Horsepower'].replace('?', np.nan)
auto_data_df['Horsepower'] = auto_data_df['Horsepower'].astype(float)

#Replace numerical data with string data for country of origin
auto_data_df['Origin'] = auto_data_df['Origin'].replace({1: 'USA', 2: 'Asia', 3: 'Europe'})

#Check data
auto_data_df.head()
```

```python
cyl_counts = auto_data_df['Cylinders'].value_counts()
cyl_counts.plot.bar()
plt.title('Cylinder Distribution')
plt.xlabel('Number of Cylinders')
plt.ylabel('Number of Cars')
plt.show()
```

```python
#set the Seaborn theme
sns.set_style('darkgrid')

#plot the data
auto_data_df.plot.scatter(x='Horsepower', y='Weight')
plt.title('Horsepower vs Weight')
plt.xlabel('Horsepower')
plt.ylabel('Weight')
plt.show()
```

### Question: How does the distribution of MPG vary between cars made in different regions of the world?

To answer this, we can use a couple of different visualizaions. First, a box plot using stripplot:

```python
# Create box plot with stripplot
sns.set_style('darkgrid')
sns.boxplot(x ='Origin', y = 'MPG', data = auto_data_df)
sns.stripplot(x ='Origin', y = 'MPG', data= auto_data_df, jitter=True, size=3)

# Add labels and title
plt.xlabel('Region of Origin')
plt.ylabel('Miles per Gallon (MPG)')
plt.title('Distribution of MPG by Region of Origin')

# Show the plot
plt.show()
```

Personally, I don't like the way that looks and I even find it a bit confusing. Let's try somethng else. This is the same data but in a violin plot, which I think looks way cooler:

```python
#Set the theme
sns.set_style('darkgrid')

# Create violin plot
sns.violinplot(x='Origin', y='MPG', data=auto_data_df)

# Add labels and title
plt.xlabel('Region of Origin')
plt.ylabel('Miles per Gallon (MPG)')
plt.title('Distribution of MPG by Region of Origin')

# Show the plot
plt.show()
```
